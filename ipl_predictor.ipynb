```python
total_score_df = delivery.groupby(["match_id","inning"]).sum()["total_runs"].reset_index()
```

```python
total_score_df = total_score_df[total_score_df["inning"]==1]
```

```python
match_df = match.merge(total_score_df[["match_id","total_runs"]],left_on ="id",right_on ="match_id")
```

```python
match_df.columns
```

```python
match_df["team1"].unique()
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians',
       
         'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
       
       'Delhi Capitals']
```

```python
change_team = {'Delhi Daredevils':'Delhi Capitals',
               'Deccan Chargers': 'Sunrisers Hyderabad'
}
```

```python
match_df["team1"]=match_df["team1"].replace(change_team)
match_df["team2"]=match_df["team2"].replace(change_team)
```

```python
match_df=match_df[match_df["team1"].isin(teams)]
match_df=match_df[match_df["team2"].isin(teams)]
```

```python
match_df["team1"].unique()
```

```python
match_df['city'].unique()
```

```python
match_df.head(2)
```

```python
match_df = match_df[match_df["dl_applied"]==0]
```

```python
match_df = match_df[["match_id","city","winner","total_runs"]]
```

```python
delivery_df = match_df.merge(delivery,on="match_id")
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df["inning"]==2]
```

```python
delivery_df.shape
```

```python
delivery_df["current_score"]=delivery_df.groupby("match_id")["total_runs_y"].cumsum()
```

```python
delivery_df
```

```python
delivery_df["runs_left"] = delivery_df["total_runs_x"] - delivery_df["current_score"]
```

```python
delivery_df.head(2)
```

```python
delivery_df["delivery_left"] = 126-(delivery_df["over"]*6 + delivery_df["ball"])
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['delivery_left'])
delivery_df['crr'] = delivery_df['crr'].replace([np.inf, -np.inf], 0).round(2)
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['delivery_left']
delivery_df['rrr'] = delivery_df['rrr'].replace([np.inf, -np.inf], 0).round(2)
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python

final_df = delivery_df[['batting_team','bowling_team','city','runs_left','delivery_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.describe()
```

```python
final_df["rrr"].isna().sum()
```

```python
final_df = final_df.sample(frac=1)
```

```python
X = final_df.drop(columns = ["result"])
y= final_df["result"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[("step1",trf),("step2",LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test ,y_pred)
```
